```python
%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
import plotutils
```

```python
init_img = cv.imread('Data/Nice.jpg')
shape = init_img.shape[1::-1]
print(shape)
```

```python
plotutils.show('Initial', init_img, 20, shape)
```

```python
reshaped = init_img.reshape((-1, 3))
pixels = np.float32(reshaped)
```

```python
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.7)
k = 5
attempts = 10
flags = cv.KMEANS_RANDOM_CENTERS
```

```python
ret, labels, centers = cv.kmeans(pixels, k, None, criteria, attempts, flags)
```

```python
centers = np.uint8(centers)
labels = labels.flatten()
```

```python
# centroids-based img
res_img = centers[labels]
final_img = res_img.reshape(init_img.shape)
plotutils.show('Result', final_img, 25, shape)
```
