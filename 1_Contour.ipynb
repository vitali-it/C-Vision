```python
%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
import plotutils
```

```python
def cv_show(name, array):
    cv.imshow(name, array)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
### For Jupyter
def plt_show(name, array):
    plt.imshow(cv.cvtColor(array, cv.COLOR_BGR2RGB))
    plt.title(name)
    plt.show()

def plt_show_inline(name, array, num):
    ax[num].imshow(cv.cvtColor(array, cv.COLOR_BGR2RGB))
    ax[num].set_title(name)
```

```python
def show(name, array):
    cv_show(name, array)
    plt_show(name, array)
```

```python
path = 'Data/Deers.jpg'
init_img = cv.imread(path)
gray_img = cv.imread(path, 2)

blank_img = np.zeros(init_img.shape[:3], dtype='uint8')
blank_img.fill(250) ## Make it white

ret, b_img = cv.threshold(gray_img, 115, 255, cv.THRESH_BINARY)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (23, 6))
plt_show_inline('Initial', init_img, 0)
plt_show_inline('Gray', gray_img, 1)
plt_show_inline('Binary', b_img, 2)
```

```python
contours, hierarchy = cv.findContours(b_img, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
image = cv.drawContours(blank_img, contours, -1, (55, 70, 220), 1)
```

```python
show('Contours', image)
```
